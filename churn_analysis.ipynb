```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(sparse_output=False, drop='first')
pd.DataFrame(
    ohe.fit_transform(df_clientes[colunas_categoricas]),
    columns=ohe.get_feature_names_out(colunas_categoricas)
)

ohe.categories_
```

```python
ohe.get_feature_names_out()
```

```python
encoded_data = ohe.transform(df_clientes[colunas_categoricas])
```

```python
df_ohe_transformadas = pd.DataFrame(data=encoded_data, columns=ohe.get_feature_names_out(), index=df_clientes.index)
df_ohe_transformadas.head()
```

```python
df_ohe_transformadas.shape
```

```python
df_clientes = pd.concat([df_clientes, df_ohe_transformadas], axis=1)
df_clientes.head()
```

```python
df_clientes = df_clientes.drop(colunas_categoricas, axis=1)
df_clientes.head()
```

```python
X = df_clientes.drop('Churn', axis=1)
y = df_clientes[['Churn']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_treino, X_teste, y_treino, y_teste = train_test_split(
    X, 
    y, 
    test_size=.2,
    random_state=42
)

X_treino.shape, X_teste.shape, y_treino.shape, y_teste.shape
```

```python
X_treino_original = X_treino.copy()
X_teste_original = X_teste.copy()

y_treino_original = y_treino.copy()
y_teste_original = y_treino.copy()
```

```python
median_dependents = X_treino['Dependents'].median()
median_tenure = X_treino['tenure'].median()
median_total_charges = X_treino['TotalCharges'].median()
moda_payment_method = X_treino['PaymentMethod'].value_counts().index[0]

median_dependents, median_tenure, median_total_charges, moda_payment_method
```

```python
X_treino['PaymentMethod'].value_counts()
```

```python
X_treino.loc[X_treino['Dependents'].isnull(), 'Dependents'] = median_dependents
X_treino.loc[X_treino['tenure'].isnull(), 'tenure'] = median_tenure
X_treino.loc[X_treino['TotalCharges'].isnull(), 'TotalCharges'] = median_total_charges
X_treino.loc[X_treino['PaymentMethod'].isnull(), 'PaymentMethod'] = moda_payment_method


X_treino.isnull().sum()
```

```python
X_teste.loc[X_teste['Dependents'].isnull(), 'Dependents'] = median_dependents
X_teste.loc[X_teste['tenure'].isnull(), 'tenure'] = median_tenure
X_teste.loc[X_teste['TotalCharges'].isnull(), 'TotalCharges'] = median_total_charges
X_teste.loc[X_teste['PaymentMethod'].isnull(), 'PaymentMethod'] = moda_payment_method

X_teste.isnull().sum()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
```

```python
rl = LogisticRegression(max_iter=500)
```

```python
type(rl)
```

```python
valores_f1_rl = cross_val_score(estimator=rl, 
                                  X=X_treino, 
                                  y=y_treino.values.flatten(), 
                                  cv=10, # 10-fold CV
                                  scoring='f1') # f1 score porque a base est desbalanceada
valores_f1_rl
```

```python
media_f1_rl = valores_f1_rl.mean()
f'f1-score: {media_f1_rl*100}'
```

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier()
```

```python
type(rfc)
```

```python
valores_f1_rfc = cross_val_score(estimator=rfc, 
                                      X=X_treino, 
                                      y=y_treino.values.flatten(), 
                                      cv=10, # 
                                  scoring='f1')
valores_f1_rfc
```

```python
media_f1_rfc = valores_f1_rfc.mean()
f'f1-score: {media_f1_rfc*100}'
```
